```python
plt.figure(figsize=(10,6))
sns.boxplot(df['cap-color'],palette='rainbow')
```

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x='odor',data=df,palette='RdBu_r')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df = df.groupby(['gill-color'], as_index=False)['cap-color'].sum().sort_values(by='cap-color',ascending=False)
sns.barplot(x ='gill-color',y= 'cap-color',data= df)
```

```python
plt.figure(figsize=(12,8))
sns.pairplot(df)
```

```python
corr_matrix=df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix,annot=True,cmap='YlGnBu')
```
